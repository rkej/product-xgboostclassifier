```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor as gb
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing 
import numpy as np
```

```python
train = pd.read_csv("train_data.csv")
train = train.drop("key", axis = 1)
X, y = train.iloc[:,:-1],train.iloc[:,-1]

for i in range(len(y)):
    if y[i] == "Class_1":
        y[i] = 0
    if y[i] == "Class_2":
        y[i] = 1
    if y[i] == "Class_3":
        y[i] = 2
    if y[i] == "Class_4":
        y[i] = 3
    if y[i] == "Class_5":
        y[i] = 4
    if y[i] == "Class_6":
        y[i] = 5
    if y[i] == "Class_7":
        y[i] = 6
    if y[i] == "Class_8":
        y[i] = 7
    if y[i] == "Class_9":
        y[i] = 8

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
eval_set = [(X_test, y_test)]
xgb_clf = xgb.XGBClassifier(objective = "multi:softmax", max_depth=250, tree_method = 'hist', grow_policy='lossguide', normalize_type = 'forest', csv_weights=1, num_round = 1200, nfold = 5,nrounds = 10,n_estimators=15000,min_child_weight=35, sample_type='weighted',colsample_bytree=0.7,subsample=0.8,eta=0.001,seed=4)
xgb_fit = xgb_clf.fit(X_train, y_train, eval_set = eval_set, eval_metric = 'mlogloss', early_stopping_rounds =70)
```

```python
test = pd.read_csv("drive/My Drive/Colab Notebooks/CMPSC 448/xgboost/test_data.csv")
test = test.drop("key", axis = 1)
preds = xgb_fit.predict(test)
for i in range(len(preds)):
  if preds[i] == 0:
    preds[i] = "Class_1"
  if preds[i] == 1:
    preds[i] = "Class_2"
  if preds[i] == 2:
    preds[i] = "Class_3"
  if preds[i] == 3:
    preds[i] = "Class_4"
  if preds[i] == 4:
    preds[i] = "Class_5"
  if preds[i] == 5:
    preds[i] = "Class_6"
  if preds[i] == 6:
    preds[i] = "Class_7"
  if preds[i] == 7:
    preds[i] = "Class_8"
  if preds[i] == 8:
    preds[i] = "Class_9"
np.unique(preds)
```
